## Tabela Verdade


And

```python
print (10 > 2 and 4 < 6)    # V and V = True
print (10 > 2 and 4 > 6)    # V and F = False
print (10 < 2 and 4 < 6)    # F and V = False
print (10 < 2 and 4 > 6)    # F and F = False
```

Or

```python
print (10 > 2 or 4 < 6)    # V or V = True
print (10 > 2 or 4 > 6)    # V or F = True
print (10 < 2 or 4 < 6)    # F or V = True
print (10 < 2 or 4 > 6)    # F or F = False
```

Not

```python
not 10 > 2   # V -> F
```

```python
not 10 < 2  # F -> V
```
