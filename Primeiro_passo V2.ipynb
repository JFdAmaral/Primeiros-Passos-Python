## Tabela Verdade


And

```python
print (10 > 2 and 4 < 6)    # V and V = True
print (10 > 2 and 4 > 6)    # V and F = False
print (10 < 2 and 4 < 6)    # F and V = False
print (10 < 2 and 4 > 6)    # F and F = False
```

Or

```python
print (10 > 2 or 4 < 6)    # V or V = True
print (10 > 2 or 4 > 6)    # V or F = True
print (10 < 2 or 4 < 6)    # F or V = True
print (10 < 2 or 4 > 6)    # F or F = False
```

Not

```python
not 10 > 2   # V -> F
```

```python
not 10 < 2  # F -> V
```

# Estruturas Condicionais


*   Conhecer a estrutura das condicionais "if", "if/else"

*   Escrever o uso dos operadores lógicos "and", "or" e "not"


```python
##if##
#O if irá executar a ação se a condição da Tabela Verdade for seguida

if 10>5:
  print ("Verdadeiro")

if 5>10:                 #Como a condição da tabela não é cumprida, o print não executa
  print ("Verdadeiro")
```

```python
##if / else##

num1 = 10
num2 = 5

if num1 > num2:
  print ("A variável num1 é o maior número")

else:
  print ("A variável num2 é o maior número")
```

```python
## if / elif / else ##

num1 = 10
num2 = 5

if num1 > num2:
  print ("A variável num1 é o maior número")

elif num1 < num2:
  print ("A variável num2 é o maior número")

else:
  print ("Os números são iguais")
```

```python
##if / else##

num1 = input ("Digite o número 1: ")
num2 = 10

num1_convert = int (num1)

if num1_convert > num2:
  print ("A variável num1 é o maior número")

elif num1_convert < num2:
  print ("A variável num2 é o maior número")

else:
  print ("Os números são iguais")
```

Condicional com Valores Lógicos

```python
num1 = input ("Digite o número 1: ")
num2 = 10
num3 = 20

num1_convert = int (num1)

if (num1_convert > num2) and (num1_convert > num3):
  print ("O num1 é o maior número")

else:
  print ("O num1 não é o maior número")
```
