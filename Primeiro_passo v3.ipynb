```python
print (10 > 2 or 4 < 6)    # V or V = True
print (10 > 2 or 4 > 6)    # V or F = True
print (10 < 2 or 4 < 6)    # F or V = True
print (10 < 2 or 4 > 6)    # F or F = False
```

# Estruturas Condicionais


*   Conhecer a estrutura das condicionais "if", "if/else"

*   Escrever o uso dos operadores lógicos "and", "or" e "not"


```python
##if##
#O if irá executar a ação se a condição da Tabela Verdade for seguida

if 10>5:
  print ("Verdadeiro")

if 5>10:                 #Como a condição da tabela não é cumprida, o print não executa
  print ("Verdadeiro")
```

```python
##if / else##

num1 = 10
num2 = 5

if num1 > num2:
  print ("A variável num1 é o maior número")

else:
  print ("A variável num2 é o maior número")
```

```python
## if / elif / else ##

num1 = 10
num2 = 5

if num1 > num2:
  print ("A variável num1 é o maior número")

elif num1 < num2:
  print ("A variável num2 é o maior número")

else:
  print ("Os números são iguais")
```

```python
##if / else##

num1 = input ("Digite o número 1: ")
num2 = 10

num1_convert = int (num1)

if num1_convert > num2:
  print ("A variável num1 é o maior número")

elif num1_convert < num2:
  print ("A variável num2 é o maior número")

else:
  print ("Os números são iguais")
```

# Estruturas de Repetição

For
Executar comandos, enquanto percorre os itens de uma coleção

Sintaxe:


* ">>> for < variável > in < coleção >:"
* ">>> << comandos >>"

Formas de repetição
• Simples
• Range
• Enumerate

```python
#For Range

for contador in range (11):
  print (contador)


```

```python
   #For Range

for contador in range(1000):
  print (contador)
  if contador == 4:
      break
```

```python
lista = ["Júlia", "Naty"]
```

```python
# Para For simples, apenas printando valores

lista = ["Júlia", "Naty"]

for val in lista:
  print (val)
```

```python
# Para For Enumerate, enumerando valores

lista = ["Júlia", "Naty"]

for i, val in enumerate (lista):
  print (i, val)
```

For - Operadores Compostos

Operadores Compostos
Muito utilizado em laços de repetição

Código mais reduzido

```python
lista2 = [2, 3, 4]

for val in lista2:
  val += 1               #Valor final soma 1
  print (val)
```

```python
lista2 = [2, 3, 4]

for val in lista2:
  val *= 2               #Válido para multiplicação
  print (val)
```

While Estrutura de repetição
Executar comandos enquanto uma condição é verdadeira

Sintaxe:
*   ">>> while < condição>:"
*   ">>> << comandos>> >>>"

e


*   "while < condição>:"
*   ">>> << comandos>>"
*   ">>> else:"
*   ">>> << comandos>>"

```python
#While

val = 1

while (val < 6):
  val *=2
  print (val)
```

```python
#While com Else

val = 9

while val < 6:
  val *=2             ## Para um valor maior que 6, o código não executa o loop por fugir da condição
else:
  print ("O valor de val é ", val)
```

```python
#While com Else

val = 3

while val < 6:
  val *=2             ## Para um valor menor que 6, o código executa o loop
else:
  print ("O valor de val é ", val)
```

```python
# Organizando dados de feedback de clientes
# Lista de feedbacks
feedbacks = ["Ótimo serviço!", "Produto excelente", "Atendimento rápido"]

# Dicionário com informações de um cliente
cliente = {    "nome": "Maria",    "idade": 28,    "feedback": "Muito satisfeita" }

# Imprimir os feedbacks da lista
print("Feedbacks dos clientes:")
for feedback in feedbacks:
  print("- " + feedback)

# Imprimir informações do cliente
  print("\nInformações do cliente:")   #\n é quebra de linha
  print("Nome:", cliente["nome"])
  print("Idade:", cliente["idade"])
  print("Feedback:", cliente["feedback"])
```

```python
# Lista de notas dos alunos
notas = [85, 92, 78, 90, 88]

# Calcula a média das notas
soma = 0
for nota in notas:
    soma += nota
media = soma / len(notas)      #\len serve para indicar o tamanho do objeto que você está trabalhando
                               #O len(Lista) será igual o nº valores dentro dessa lista

# Classifica os alunos com base na média
alunos_aprovados = []
alunos_reprovados = []

for nota in notas:
  if nota >= media:
     alunos_aprovados.append(nota)
  else:
    alunos_reprovados.append(nota)

# Imprime os resultados

print("Notas dos alunos:", notas)
print("Média das notas:", media)
print("Alunos aprovados:", alunos_aprovados)
print("Alunos reprovados:", alunos_reprovados)
```
